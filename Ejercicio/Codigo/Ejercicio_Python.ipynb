```python
# Hacer subset para tienda nueva y vieja
tiendasNuevas = df[df['Tipo Tienda'] == 'Nueva']
tiendasViejas = df[df['Tipo Tienda'] == 'Vieja']
```

Evaluar si las tiendas Nuevas mejoran en comparación con las Viejas. Se aplico una nueva forma
en la cual siempre se obtiene descuento extra en el ticket de compra. Se busca ver si en efecto,
esta estrategia comercial promueve un mayor ticket promedio de venta, venta neta y en
comparación con las otras tiendas se obtienen mejores beneficios.

Cada fila es una venta, la columna venta fue lo que se gastó y la de descuento es lo que se le descontó. Numero de productos es la cantidad de productos que se llevo la persona y tienda se repite, pues muchos clientes compran en la misma tienda.

```python
# Calcular el ticker promedio de venta para las tiendas nuevas y viejas y compararlos

# Calcular el ticket promedio de venta para tiendas nuevas
venta_promedio_nueva = tiendasNuevas['Venta'].mean()

# Calcular el ticket promedio de venta para tiendas viejas
venta_promedio_vieja = tiendasViejas['Venta'].mean()

# Calculamos la diferencia entre ticket promedio de tiendas nuevas y viejas
diferencia_ticker_nv = venta_promedio_nueva - venta_promedio_vieja

# Comparar los ticket promedio de venta para tiendas nuevas y viejas
print('Ticket promedio de tiendas nuevas: ', venta_promedio_nueva)
print('Ticket promedio de tiendas viejas: ', venta_promedio_vieja)
print('Diferencia entre ticket promedio de tiendas nuevas y viejas: ', diferencia_ticker_nv)
```

```python
# Calcular la venta neta para cada tipo de tienda:

# Calcular la venta neta para tiendas nuevas
venta_neta_nuevas = tiendasNuevas['Venta'] - tiendasNuevas['Descuento']

# Calcular la venta neta para tiendas viejas
venta_neta_viejas = tiendasViejas['Venta'] - tiendasViejas['Descuento']

# Calcular la diferencia entre venta neta de tiendas nuevas y viejas
diferencia_venta_nv = venta_neta_nuevas.sum() - venta_neta_viejas.sum()

# Comparar la venta neta para tiendas nuevas y viejas
print('Venta neta de tiendas nuevas: ', venta_neta_nuevas.sum())
print('Venta neta de tiendas viejas: ', venta_neta_viejas.sum())
print('Diferencia entre venta neta de tiendas nuevas y viejas: ', diferencia_venta_nv)
```

```python
# Cuantas transacciones hubo por tienda

# Contar transacciones por tienda
transacciones_CDMX =len(df[df['Tienda'] == 'CDMX – 35']['Tienda'])
transacciones_Chihuahua =len(df[df['Tienda'] == 'Chihuahua – 71']['Tienda'])
transacciones_Playa =len(df[df['Tienda'] == 'Playa del Carmen – 100']['Tienda'])


print("Las transacciones por tienda CDMX son: ", transacciones_CDMX)
print("Las transacciones por tienda Chihuahua son: ", transacciones_Chihuahua)
print("Las transacciones por tienda Playa del Carmen son: ", transacciones_Playa)
```

```python
# Agrupar transacciones por tienda y calcular la venta total, el descuento total y el ticket promedio de venta para cada tienda:

# Agrupar transacciones por tienda
agrupados_tienda = df.groupby('Tienda').sum()

# Calcular el ticket promedio de venta para cada tienda
ticket_promedio = []

# CDMX
ticket_promedio_cdmx = agrupados_tienda['Venta']['CDMX – 35'] / transacciones_CDMX
ticket_promedio.append(ticket_promedio_cdmx)

# Chihuahua
ticket_promedio_chihuahua = agrupados_tienda['Venta']['Chihuahua – 71'] / transacciones_Chihuahua
ticket_promedio.append(ticket_promedio_chihuahua)

# Playa del Carmen
ticket_promedio_playa = agrupados_tienda['Venta']['Playa del Carmen – 100'] / transacciones_Playa
ticket_promedio.append(ticket_promedio_playa)

# Agregar columna ticket promedio a la tabla agrupados_tienda
agrupados_tienda['Ticket Promedio'] = ticket_promedio


print(agrupados_tienda)
```

Los resultados que se obtuvieron indican que la tienda nueva (Playa del Carmen), tiene un número de transacciones similar al de las tiendas viejas (CDMX y Chihuahua). Sin embargo, el ticket promedio de la tienda nueva es significativamente mayor que el de las tiendas viejas, lo que indica que la tienda nueva está obteniendo mayores ingresos por transacción.

La venta neta de la tienda nueva también es mayor que la de las tiendas viejas, con una diferencia de 875.0483. Esto sugiere que la estrategia de aplicar un descuento extra en las tiendas nuevas está siendo efectiva en términos de aumentar los ingresos totales.

Esto nos dice que la estrategia de aplicar un descuento extra al aperturar tiendas nuevas si es una estrategia efectiva para aumentar los ingresos totales y el ticket promedio.

# 5.- Hacer una gráfica de barras para las 3 tiendas con la primera barra de la Venta Neta, y la segunda barra del descuento, con lo que ambas stackeadas dan la Venta.

```python
# Ventas netas y descuentos por tienda

# CDMX
venta_neta_CDMX = df[df['Tienda'] == 'CDMX – 35']['Venta'].sum() - df[df['Tienda'] == 'CDMX – 35']['Descuento'].sum()
descuento_CDMX = df[df['Tienda'] == 'CDMX – 35']['Descuento'].sum()

# Chihuahua
venta_neta_Chihuahua = df[df['Tienda'] == 'Chihuahua – 71']['Venta'].sum() - df[df['Tienda'] == 'Chihuahua – 71']['Descuento'].sum()
descuento_Chihuahua = df[df['Tienda'] == 'Chihuahua – 71']['Descuento'].sum()

# Playa del Carmen
venta_neta_Playa = df[df['Tienda'] == 'Playa del Carmen – 100']['Venta'].sum() - df[df['Tienda'] == 'Playa del Carmen – 100']['Descuento'].sum()
descuento_Playa = df[df['Tienda'] == 'Playa del Carmen – 100']['Descuento'].sum()

print("La venta neta de CDMX es: ", venta_neta_CDMX, "Y el descuento es de: ", descuento_CDMX, 'La suma de ambos es: ', venta_neta_CDMX + descuento_CDMX)
print("La venta neta de Chihuahua es: ", venta_neta_Chihuahua, "Y el descuento es de: ", descuento_Chihuahua, 'La suma de ambos es: ', venta_neta_Chihuahua + descuento_Chihuahua)
print("La venta neta de Playa del Carmen es: ", venta_neta_Playa, "Y el descuento es de: ", descuento_Playa, 'La suma de ambos es: ', venta_neta_Playa + descuento_Playa)
```

```python
# Hacer una gráfica de barras para las 3 tiendas con la primera barra de la Venta Neta, y la segunda barra del descuento, con lo que ambas stackeadas dan la Venta.

# Crear un dataframe con las ventas netas y descuentos por tienda
ventas = pd.DataFrame({'Venta Neta': [venta_neta_CDMX, venta_neta_Chihuahua, venta_neta_Playa], 
                        'Descuento': [descuento_CDMX, descuento_Chihuahua, descuento_Playa]}, 
                        index=['CDMX', 'Chihuahua', 'Playa del Carmen'])

# Crear la gráfica de barras
ventas.plot.bar(stacked=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(100)
y = 100 + x * 2 + np.random.rand(100) * 50
plt.plot(x, y, 'ro')

from sklearn.linear_model import LinearRegression

# Crear un modelo de regresión lineal
model = LinearRegression()

# Ajustar los datos a la regresión lineal
model.fit(x.reshape(-1, 1), y)

# Calcular los coeficientes y el intercepto
print('Coeficiente: ', model.coef_[0])
print('Intercepto: ', model.intercept_)

# Graficar la recta de regresión
plt.plot(x, model.predict(x.reshape(-1, 1)), color='blue')
plt.show()
```

El modelo de regresión lineal simple, sugiere que existe una relación lineal positiva entre las variables x/y, es decir, a medida que x aumenta, y también aumenta.
El coeficiente de 2.064865701557537 indica que por cada unidad de cambio en x, y cambia en 2.064865701557537 unidades, lo cual significa que existe una relación directa entre x/y.
